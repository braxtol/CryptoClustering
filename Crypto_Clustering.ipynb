```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
df_market_data_scaled = StandardScaler().fit_transform(df_market_data[['price_change_percentage_24h', 'price_change_percentage_7d',
       'price_change_percentage_14d', 'price_change_percentage_30d',
       'price_change_percentage_60d', 'price_change_percentage_200d',
       'price_change_percentage_1y']])
```

```python
# Create a DataFrame with the scaled data
df_market_scaled=pd.DataFrame(df_market_data_scaled, 
                             columns=['price_change_percentage_24h', 'price_change_percentage_7d',
                                       'price_change_percentage_14d', 'price_change_percentage_30d',
                                       'price_change_percentage_60d', 'price_change_percentage_200d',
                                       'price_change_percentage_1y'])

# Copy the crypto names from the original data

df_market_scaled['coin_id']=df_market_data.index

# Set the coinid column as index
df_market_scaled=df_market_scaled.set_index('coin_id')

# Display sample data
df_market_scaled.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k=list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for i in k:
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
    k_model= KMeans(n_clusters=i, random_state=42)
# 2. Fit the model to the data using `df_market_data_scaled`
    k_model.fit(df_market_data_scaled)
# 3. Append the model.inertia_ to the inertia list
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {'k_original': k, 'inertia': inertia}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_plot = elbow_df.hvplot.line(
    x = 'k_original', 
    y = 'inertia', 
    title = 'Elbow Curve', 
    xticks = k
)
elbow_plot
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** If I was to choose the best value for `k` I would choose `k = 4` because that is where the elbow of the graph is.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters = 4)
```

```python
# Fit the K-Means model using the scaled data
model.fit(df_market_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
predictions = model.predict(df_market_scaled)

# Print the resulting array of cluster values.
predictions
```

```python
#This drives Henry bonkers, so I am going to do it.
print(predictions)
```

```python
# Create a copy of the DataFrame
df_market_scaled_predictions = df_market_scaled.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_market_scaled_predictions['Predictions'] = predictions


# Display sample data
df_market_scaled_predictions.head(8)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

predictions_scatter_plot = df_market_scaled_predictions.hvplot.scatter(
    title = 'Scatter Plot by Crypto Currency Segment where k = 4',
    x = 'price_change_percentage_24h',
    y = 'price_change_percentage_7d',
    xlabel = '24 Hr Price Change %',
    ylabel = '7-Day Price Change %',
    by = 'Predictions',
    hover_cols = ['coin_id'],
    frame_width = 700,
    frame_height = 500
)

predictions_scatter_plot
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components = 3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
coin_pca_data = pca.fit_transform(df_market_scaled)
# View the first five rows of the DataFrame. 
coin_pca_data[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
variance = pca.explained_variance_ratio_

print(f'PC1: {variance[0]:.2f}%\nPC2: {variance[1]:.2f}%\nPC3: {variance[2]:.2f}%\nThe primary components hold a total of{sum(variance) * 100: .2f}% of the original data.')
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance is a sum of 89.50% of the original data.

```python
# Create a new DataFrame with the PCA data.

# Creating a DataFrame with the PCA data
df_market_data_pca = pd.DataFrame(coin_pca_data, columns=['PC1', 'PC2', 'PC3'])

# Copy the crypto names from the original data
df_market_data_pca['coin_id'] = df_market_scaled.index

# Set the coinid column as index
df_market_data_pca = df_market_data_pca.set_index('coin_id')


# Display sample data
df_market_data_pca.head(8)
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
pca_k_values = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
pca_inertia = []

# Create a for loop to compute the inertia with each possible value of k
for k in pca_k_values: 
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
    model = KMeans(n_clusters=k)
# 2. Fit the model to the data using `df_market_data_pca`
    model.fit(df_market_data_pca)
# 3. Append the model.inertia_ to the inertia list
    pca_inertia.append(model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
pca_elbow_data = {
    'k': pca_k_values,
    'Inertia': pca_inertia
}
# Create a DataFrame with the data to plot the Elbow curve
df_pca_elbow = pd.DataFrame(pca_elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
pca_elbow_plot = df_pca_elbow.hvplot(x='k', y='Inertia', title='PCA K Values Elbow Curve', frame_width=800, frame_height=400)
pca_elbow_plot
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** The best value for `k` is `4` because it is the best value found using the PCA data.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No, it does not differ from the best k value found using the original data.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
pca_model = KMeans(n_clusters=4)
```

```python
# Fit the K-Means model using the PCA data
pca_model.fit(df_market_data_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
coin_segments_pca = pca_model.predict(df_market_data_pca)

# Print the resulting array of cluster values.
print(coin_segments_pca)
```

```python
# Create a copy of the DataFrame with the PCA data
market_data_pca_predictions = df_market_data_pca.copy()

# Add a new column to the DataFrame with the predicted clusters
market_data_pca_predictions['CoinClusters'] = coin_segments_pca

# Display sample data
market_data_pca_predictions.head(8)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

pca_predictions_scatter_plot = market_data_pca_predictions.hvplot.scatter(
    title='Scatter Plot by Crypto Currency Segment Using Primary Clusters where k = 4',
    x='PC1',
    y='PC2',
    by='CoinClusters',
    hover_cols=['coin_id'],
    frame_width=700,
    frame_height=500
)

pca_predictions_scatter_plot
```
